```python
import pickle

# Save kiya hua model load karein
with open('./models/rf_model.pkl', 'rb') as model_file:
    loaded_model = pickle.load(model_file)

print("âœ… Model successfully loaded!")

# Prediction Example
sample_data = [[523400, 15, 20, 120.5, 5.6, 320, 30000, 6, 1, 0]]  # Example input
prediction = loaded_model.predict(sample_data)
print("Prediction:", prediction[0])
```

```python
# Sample network traffic data
traffic_samples = np.array([
    [9909819299, 5, 000, 80090, 0, 50023420, 99, 1, 0, 50000],
    [80000, 2, 180, 75234200, 2, 4000, 850000, 0, 1, 4800],  
    [946530, 50, 323420, 100, 10, 200, 100, 8, 3, 900],     
    [50000, 14, 300, 6, 1000, 7000, 62000, 0, 1000, 490000],
    [90000, 30, 40, 150, 15, 250, 12000, 5, 1, 1000] 
])

# Model se prediction lein
predictions = loaded_model.predict(traffic_samples)
probabilities = loaded_model.predict_proba(traffic_samples)[:, 1]  # Probability of fraud

# Results ko print karein
for i, (pred, prob) in enumerate(zip(predictions, probabilities)):
    if pred == 1:
        print(f"Sample {i+1}: ðŸš¨ Malicious Traffic Detected! (Prob: {prob:.2f})")
    else:
        print(f"Sample {i+1}: âœ… Safe Traffic (Prob: {prob:.2f})")
```
